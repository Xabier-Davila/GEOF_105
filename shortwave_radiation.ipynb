Calculate the shortwave radiation 

```python
#Define Variables
alpha = 40 #albedo
Qc =  50 #incoming clear-sky solar radiation at top of atmosphere
C = 0.7 #monthly mean fractional cloud cover
theta_N = 60 #noon solar elevation (degrees)
```

```python
Qs = (1-alpha) * Qc * (1 - 0.62 *C + 0.0019 * theta_N)
```

```python
print('Shortwave radiation =',Qs)
```
