# Calculate the shortwave radiation 

1. We first import libraries to be used.



```python
import numpy as np
import matplotlib.pyplot as plt
```

2. Now we define the variables 

alpha = Albedo (%) 

Qc = Incoming clear-sky radiation at top of atmosphere (W/m2)

C = monthly mean fractional cloud cover (%)

theta_N = noon solar elevation (degrees)

```python
#Define Variables

alpha = 0.06 
Qc =  340 
C = 0.7 
theta_N = 60 
```

```python
#Apply the formula
Qs = (1-alpha) * Qc * (1 - 0.62 * C + 0.0019 * theta_N)
```

```python
#print the result
print('Shortwave radiation =',Qs, '(W/m2)')
```

3. How does a change in incoming solar radiation affect the shortwave radiation?

```python
#Select maximum and minimum values for incoming solar radiation
Qc_max = 400
Qc_min = 100

#create an array between the max and min values
Qc_serie = np.linspace(Qc_max, Qc_min, 30)
```

```python
#Apply formula
Qs_serie_1 = (1-alpha) * Qc_serie * (1 - 0.62 * C + 0.0019 * theta_N)
```

```python
#plot results
plt.plot(Qc_serie, Qs_serie_1)
plt.xlabel('Qc')
plt.ylabel('Qs (W/m2)')
```

4. What about modifying the cloud cover?

```python
#Select maximum and minimum values for incoming solar radiation
C_max = 0
C_min = 1

#create an array between the max and min values
C_serie = np.linspace(C_max, C_min, 30) #30 time steps
```

```python
Qs_serie_2 = (1-alpha) * Qc * (1 - 0.62 * C_serie + 0.0019 * theta_N)
```

```python
#plot results
plt.plot(C_serie, Qs_serie_2)
plt.xlabel('Cloud cover (%)')
plt.ylabel('Qs (W/m2)')
```

5. What happens when these two factor vary simultaneously?

```python
Qs_serie_3 = (1-alpha) * Qc_serie * (1 - 0.62 * C_serie + 0.0019 * theta_N)
```

```python
#plot results
plt.plot(Qs_serie_3)
plt.xlabel('Time Step') #Increasing cloud cover and increasing radiation
plt.ylabel('Qs')
```
