# Calculate the longwave radiation 

1. Import libraries 



```python
import numpy as np
import matplotlib.pyplot as plt
```

2. We define the variables 

```python
#Define Variables

alpha = 0.06 #albedo (%)
Qc =  340 #incoming clear-sky solar radiation at top of atmosphere (W/m2)
C = 0.7 #monthly mean fractional cloud cover ()
theta_N = 60 #noon solar elevation (degrees)
```

```python
#Apply the formula
Qs = (1-alpha) * Qc * (1 - 0.62 * C + 0.0019 * theta_N)
```

```python
#print the result
print('Shortwave radiation =',Qs, '(W/m2)')
```

3. How does a change in incoming solar radiation affect the shortwave radiation?

```python
#Select maximum and minimum values for incoming solar radiation
Qc_max = 400
Qc_min = 100

#create an array between the max and min values
Qc_serie = np.linspace(Qc_max, Qc_min, 30)
```

```python
#Apply formula
Qs_serie = (1-alpha_serie) * Qc * (1 - 0.62 * C + 0.0019 * theta_N)
```

```python
#plot results
plt.plot(Qc_serie, Qs_serie)
plt.xlabel('Qc')
plt.ylabel('Qs')
```

3. What about modifying the cloud cover?

```python
#Select maximum and minimum values for incoming solar radiation
C_max = 0
C_min = 1

#create an array between the max and min values
C_serie = np.linspace(C_max, C_min, 30) #30 time steps
```

```python
Qs_serie = (1-alpha) * Qc * (1 - 0.62 * C_serie + 0.0019 * theta_serie)
```

```python
#plot results
plt.plot(C_serie, Qs_serie)
plt.xlabel('Cloud cover (%)')
plt.ylabel('Qs')
```

4. What happens when these two factor vary simultaneously?

```python
Qs_serie = (1-alpha) * Qc_serie * (1 - 0.62 * C_serie + 0.0019 * theta_serie)
```

```python
#plot results
plt.plot(Qs_serie)
plt.xlabel('Time Step')
plt.ylabel('Qs')
```
